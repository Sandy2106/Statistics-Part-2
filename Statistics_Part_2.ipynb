# Statistics Part 2

1. What is hypothesis testing in statistics?
- Hypothesis testing is a statistical method used to make decisions about a population based on sample data. It involves testing an assumption (hypothesis) about a parameter and using sample data to determine the likelihood that this assumption is true.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
- Null hypothesis (H₀): A statement that there is no effect or no difference. It is the default or starting assumption.

- Alternative hypothesis (H₁ or Ha): A statement that contradicts the null hypothesis, indicating the presence of an effect or a difference.

3. What is the significance level in hypothesis testing, and why is it important?
- The significance level (α) is the probability of rejecting the null hypothesis when it is actually true. Common levels are 0.05, 0.01, or 0.10. It defines the threshold for statistical significance and helps control the risk of a Type I error.

4. What does a P-value represent in hypothesis testing?
- The P-value is the probability of obtaining test results at least as extreme as the observed results, assuming that the null hypothesis is true. It measures the evidence against the null hypothesis.

5. How do you interpret the P-value in hypothesis testing?
-
 - If P-value ≤ α, reject the null hypothesis (evidence supports the alternative).

 - If P-value > α, fail to reject the null hypothesis (not enough evidence).

6. What are Type 1 and Type 2 errors in hypothesis testing?
- Type I error (False Positive): Rejecting a true null hypothesis.

- Type II error (False Negative): Failing to reject a false null hypothesis.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
- One-tailed test: Tests for deviation in one direction (e.g., greater than or less than).

- Two-tailed test: Tests for deviation in both directions (e.g., not equal to).

8. What is the Z-test, and when is it used in hypothesis testing?
- The Z-test is used when the population standard deviation is known and the sample size is large (n ≥ 30). It compares a sample mean to a population mean.

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
- Z-score = (Sample Mean − Population Mean) / (Standard Deviation / √n)
It represents how many standard deviations a data point is from the population mean.

10. What is the T-distribution, and when should it be used instead of the normal distribution?
- The T-distribution is used when the sample size is small (n < 30) and the population standard deviation is unknown. It is wider and has heavier tails than the normal distribution.

11. What is the difference between a Z-test and a T-test?
- Z-test: Used when population standard deviation is known and n is large.

- T-test: Used when population standard deviation is unknown and/or n is small.

12. What is the T-test, and how is it used in hypothesis testing?
- The T-test compares sample means to determine if there is a significant difference between groups. Types include:

  - One-sample T-test
  - Independent two-sample T-test
  - Paired sample T-test

13. What is the relationship between Z-test and T-test in hypothesis testing?
- Both are used to test hypotheses about means. The Z-test is appropriate for large samples with known population variance, while the T-test is used when the sample is small and variance is unknown.

14. What is a confidence interval, and how is it used to interpret statistical results?
- A confidence interval is a range of values likely to contain the population parameter with a certain level of confidence (e.g., 95%). It provides an estimate of uncertainty.

15. What is the margin of error, and how does it affect the confidence interval?
- The margin of error is the amount added and subtracted from the sample statistic to create the confidence interval. A larger margin of error means a wider interval, indicating more uncertainty.

16. How is Bayes' Theorem used in statistics, and what is its significance?
- Bayes' Theorem calculates the probability of an event based on prior knowledge of conditions related to the event. It's essential in updating beliefs with new evidence (used in Bayesian statistics).

17. What is the Chi-square distribution, and when is it used?
- The Chi-square distribution is used in tests of independence and goodness of fit. It’s applicable when dealing with categorical variables and frequencies.

18. What is the Chi-square goodness of fit test, and how is it applied?
- This test checks whether a sample data set fits a particular distribution. It compares observed frequencies to expected frequencies using the Chi-square statistic.

19. What is the F-distribution, and when is it used in hypothesis testing?
- The F-distribution is used primarily in analysis of variance (ANOVA) and in testing the equality of two variances. It’s skewed and only takes positive values.

20. What is an ANOVA test, and what are its assumptions?
- ANOVA (Analysis of Variance) tests whether there are significant differences between the means of three or more groups.

 - Assumptions include:
 - Independence of observations
 - Normality
 - Equal variances (homogeneity of variance)

21. What are the different types of ANOVA tests?
- One-way ANOVA: Tests one independent variable.

- Two-way ANOVA: Tests two independent variables.

- Repeated Measures ANOVA: Tests same subjects under different conditions/times.

22. What is the F-test, and how does it relate to hypothesis testing?
- The F-test compares two variances to test if they are significantly different. It is used in ANOVA and regression analysis to test model significance.

1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results.

```python
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    # Calculate sample statistics
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    # Calculate Z-score
    standard_error = population_std / np.sqrt(sample_size)
    z_score = (sample_mean - population_mean) / standard_error

    # Calculate two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Print the results
    print("Sample Mean:", sample_mean)
    print("Z-Score:", z_score)
    print("P-Value:", p_value)

    # Interpretation
    if p_value < alpha:
        print(f"Result: Reject the null hypothesis at α = {alpha}")
        print("Interpretation: The sample mean is significantly different from the population mean.")
    else:
        print(f"Result: Fail to reject the null hypothesis at α = {alpha}")
        print("Interpretation: The sample mean is not significantly different from the population mean.")

# Example usage
sample = [52, 55, 53, 54, 50, 56, 52, 54, 53, 55]
known_population_mean = 50
known_population_std = 2  # Population standard deviation (σ)

z_test(sample, known_population_mean, known_population_std)
```

2. Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python.

```python
import numpy as np
from scipy.stats import norm

# Simulate sample data
np.random.seed(0)
sample = np.random.normal(loc=52, scale=5, size=30)  # Simulate sample with mean 52
population_mean = 50
population_std = 5

# Hypothesis Testing
sample_mean = np.mean(sample)
n = len(sample)
z = (sample_mean - population_mean) / (population_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z)))

print("Sample Mean:", sample_mean)
print("Z-score:", z)
print("P-value:", p_value)
```

3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

```python
def one_sample_z_test(sample, pop_mean, pop_std):
    sample_mean = np.mean(sample)
    n = len(sample)
    se = pop_std / np.sqrt(n)
    z = (sample_mean - pop_mean) / se
    p = 2 * (1 - norm.cdf(abs(z)))
    return z, p

z, p = one_sample_z_test(sample, population_mean, population_std)
print("Z:", z, "| P-value:", p)
```

4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import matplotlib.pyplot as plt

def visualize_z_test(z_score, alpha=0.05):
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)
    critical = norm.ppf(1 - alpha/2)

    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label='Normal Distribution')
    plt.fill_between(x, y, where=(x <= -critical) | (x >= critical), color='red', alpha=0.5, label='Rejection Region')
    plt.axvline(z_score, color='blue', linestyle='--', label=f'Z = {z_score:.2f}')
    plt.title("Two-tailed Z-test Decision Region")
    plt.legend()
    plt.show()

visualize_z_test(z)
```

5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
def visualize_errors(mu_0, mu_1, sigma, n, alpha=0.05):
    se = sigma / np.sqrt(n)
    x = np.linspace(mu_0 - 4*se, mu_1 + 4*se, 1000)
    h0 = norm.pdf(x, mu_0, se)
    h1 = norm.pdf(x, mu_1, se)

    z_critical = norm.ppf(1 - alpha)
    threshold = mu_0 + z_critical * se

    plt.plot(x, h0, label='H0: μ=μ0')
    plt.plot(x, h1, label='H1: μ=μ1')
    plt.fill_between(x, h0, where=(x >= threshold), color='red', alpha=0.5, label='Type I Error (α)')
    plt.fill_between(x, h1, where=(x < threshold), color='blue', alpha=0.3, label='Type II Error (β)')
    plt.axvline(threshold, color='black', linestyle='--')
    plt.legend()
    plt.title("Type I and Type II Error Visualization")
    plt.show()

visualize_errors(mu_0=50, mu_1=52, sigma=5, n=30)
```

9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
def confidence_interval(sample, confidence=0.95):
    mean = np.mean(sample)
    se = np.std(sample, ddof=1) / np.sqrt(len(sample))
    margin = norm.ppf((1 + confidence) / 2) * se
    return mean - margin, mean + margin

ci = confidence_interval(sample)
print("Confidence Interval (95%):", ci)
```

```python
def margin_of_error(sample, confidence=0.95):
    se = np.std(sample, ddof=1) / np.sqrt(len(sample))
    return norm.ppf((1 + confidence) / 2) * se

moe = margin_of_error(sample)
print("Margin of Error (95%):", moe)
```

11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
def bayes_theorem(prior_A, likelihood_B_given_A, prior_not_A, likelihood_B_given_not_A):
    numerator = likelihood_B_given_A * prior_A
    denominator = (likelihood_B_given_A * prior_A) + (likelihood_B_given_not_A * prior_not_A)
    return numerator / denominator

# Example
P_A = 0.01  # Disease
P_B_given_A = 0.95  # Positive test if diseased
P_not_A = 0.99
P_B_given_not_A = 0.05  # False positive

posterior = bayes_theorem(P_A, P_B_given_A, P_not_A, P_B_given_not_A)
print("P(Disease | Positive Test):", posterior)
```

16. Implement an F-test using Python to compare the variances of two random samples.

```python
def f_test(sample1, sample2):
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)
    f = var1 / var2
    dfn = len(sample1) - 1
    dfd = len(sample2) - 1
    from scipy.stats import f as f_dist
    p = 2 * min(f_dist.cdf(f, dfn, dfd), 1 - f_dist.cdf(f, dfn, dfd))
    return f, p

s1 = np.random.normal(10, 2, 30)
s2 = np.random.normal(10, 5, 30)
f_stat, p_val = f_test(s1, s2)
print("F-statistic:", f_stat, "| P-value:", p_val)
```

19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

```python
from scipy.stats import shapiro, levene

def check_anova_assumptions(*groups):
    # Normality
    for i, g in enumerate(groups, 1):
        stat, p = shapiro(g)
        print(f"Group {i} Normality P-value: {p:.4f} -> {'OK' if p > 0.05 else 'Fail'}")

    # Homogeneity of variance
    stat, p = levene(*groups)
    print(f"Equal Variance (Levene’s) P-value: {p:.4f} -> {'OK' if p > 0.05 else 'Fail'}")

check_anova_assumptions(group1, group2, group3)
```

24.  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.
